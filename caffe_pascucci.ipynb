# 정적 웹 페이지 크롤링 실습 - 파스쿠찌 커피

https://www.caffe-pascucci.co.kr/store/storeList.asp

```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
import datetime
```

```python
def pascucci_store(result):
    for page in range(1, 10):
        #주차+와이파이 가능한 전국의 매장
        pascucci_url = 'https://www.caffe-pascucci.co.kr/store/storeList.asp?searchSmoking=&searchParking=Y&searchInternet=Y&searchNewStore=&search24Store=&searchCatering=&searchDrivethrough=&searchKey=&page=%d&searchRegion=&searchGugun=&searchValue=' %page
        html = urllib.request.urlopen(pascucci_url)
        pascucci_soup = BeautifulSoup(html, 'html.parser')
        tbody = pascucci_soup.find('tbody')
        
        for store in tbody.find_all('tr'):
            # 예외처리
            if len(store) < 3:
                break
            
            
            store_td = store.find_all('td')
            
            store_name = store_td[1].string # 점포명
            store_region = store_td[2].string # 지역
            store_address = store_td[4].p.string # 주소
            store_call = store_td[5].span.string # 전화번호
            
            result.append([store_name] + [store_region] + [store_address] + [store_call])
            print(result)
    return
```

```python
def main():
    result = []
    print('=====파스쿠치 크롤링 작업 시작====')
    pascucci_store(result)
    print('=====파스쿠치 크롤링 작업 종료====')
    pascucci_data = pd.DataFrame(result, columns = {'store', 'region', 'address', 'call'})
    pascucci_data.to_csv('./pascucci.csv', encoding='cp949', mode='w', index=True)
    del result[:]
```

```python
main()
```
