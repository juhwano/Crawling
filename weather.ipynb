# 기상청 날씨 불러오기

https://www.weather.go.kr/weather/observation/currentweather.jsp

```python
# 크롤링 기본 세팅
import requests # 서버접속을 위한 lib
from bs4 import BeautifulSoup # 크롤링을 위한 lib
import pandas as pd
```

```python
source = requests.get('https://www.weather.go.kr/weather/observation/currentweather.jsp')
soup = BeautifulSoup(source.content, 'html.parser')
```

```python
table = soup.find('table', {'class' : 'table_develop3'})
data = []
```

```python
table
```

```python
# 출력을 위한 문자열 꾸미기
print('='*30)
print('\t 오늘의 날씨')
print('='*30)

# code 0
for tr in table.find_all('tr'):
    tds = list(tr.find_all('td'))
    for td in tds:
        if td.find('a'):
            location = td.find('a').text # 위치
            temp = tds[5].text # 기온
            humidity = tds[9].text # 습도
#             print(location, temp, humidity)
            print('{0:<7} {1:<7} {2:<7}'.format(location, temp, humidity))
            # {0:<10}이 뜻하는 것은 {0} 값을 ":<10" 10자리로 표현할건데 왼쪽 정렬을 할 것이라 라는 뜻
            # {1:>5} 가 뜻하는 것은 {1} 값을 ":>5" 5자리로 표현할건데 오른쪽 정렬을 할 것이다라는 뜻
            # 왼쪽 정렬에는 기호 < 를 사용하고, 오른쪽 정렬에는 >, 가운데 정렬에는 ^
            
            data.append([location, temp, humidity])
            
    

# 출력 종료
print('='*30)
print('\t 출력 종료')
print('='*30)
```
